```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt("data.csv", delimiter=',')
```

```python
x = data[:,0]
y = data[:,1]
```

```python
plt.plot(x,y,'-r')
plt.xlabel("x")
plt.ylabel("sinyal")
plt.show()
```

```python
plt.plot(x,y, 'or')
plt.xlabel("x")
plt.ylabel("sinyal")
plt.show()
```

```python
plt.stem(x,y, use_line_collection = True)
plt.xlabel("x")
plt.ylabel("sinyal")
plt.show()
```
